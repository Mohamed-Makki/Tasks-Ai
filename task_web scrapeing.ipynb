```python
#dowenload image
import requests
from bs4 import BeautifulSoup
weburl="https://www.alahlyegypt.com/ar/football/team"
website=requests.get(weburl).content
soup=BeautifulSoup(website,'lxml')
print(soup)
imgs=soup.find_all('img')
print(imgs)
imgs=imgs[3:-1]
for img in imgs:
    link=img.attrs.get('src')
    print(link)
    src=requests.get(link).content
    filename = r"paper" + link[link.rfind("/"):]
    with open(filename,'wb') as file:
        file.write(src)
```

```python
#scraping 5 article
import requests
from bs4 import BeautifulSoup
import csv
website=requests.get("https://coreyms.com/").text
soup=BeautifulSoup(website,"lxml")
articles=soup.find_all("article")
with open("articlecsv","w",encoding='utf-8') as file:
    filecsv=csv.writer(file)
    filecsv.writerow(['articles.content'])
    for article in articles:
        if article != articles[5]:
            filecsv.writerow([article])
        else:
            filecsv.close()
            break
```
